```python
import requests, re
```

```python
res = requests.get('https://unicode.org/Public/emoji/13.0/emoji-test.txt')
```

```python
emojis = {c for x in {
    re.sub(r' .*', '', re.sub('.*#', '', line).strip())
    for line in
    res.text.splitlines()
    if ' E' in line
} if x != '' for c in x if not re.search('[A-Za-z 0-9,:;\.\(\)]', c)}
emojis
```

```python
with open('../emojis.txt', 'w') as f:
    f.write(''.join(sorted(emojis)))
```
